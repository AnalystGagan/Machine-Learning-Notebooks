## Matplotlib Example Workflow


```python
# 1.Import matplotlib and get it ready for plotting in jupyter
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 2.Prepare Data
x  = [1,2,3,4,5]
y = [11,22,33,44,55]  

#3.Setup a plot
fig, ax = plt.subplots(figsize = (10,10)) #(10,10) height and width

#4. Plot the data
ax.plot(x,y)

#5. customize plot
ax.set(title = "simple plot",xlabel = "X-axis",ylabel = "Y=axis")

#6. Save and show (you save the whole figure)
fig.savefig("firstplot.png")
```
