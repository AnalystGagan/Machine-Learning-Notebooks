## Matplotlib Example Workflow


```python
# 1.Import matplotlib and get it ready for plotting in jupyter
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# 2.Prepare Data
x  = [1,2,3,4,5]
y = [11,22,33,44,55]  

#3.Setup a plot
fig, ax = plt.subplots(figsize = (10,10)) #(10,10) height and width

#4. Plot the data
ax.plot(x,y)

#5. customize plot
ax.set(title = "simple plot",xlabel = "X-axis",ylabel = "Y=axis")

#6. Save and show (you save the whole figure)
fig.savefig("firstplot.png")
```

### pyplot vs matplotlib vs OO method

* When plotting somthing quickly, it is ok to use pyplot method
* When plotting somthing advance, use OO method


```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# pyplot method aka directly from pandas dataframe
over_50.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
# oo method mixed with pyplot method

fig, ax = plt.subplots(figsize = (10,6)) #ax object
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax = ax);

#ax argument in plot( ) make sure that this data get plotted on ax objected that we have created 
# ax.set_xlim ([45,100])
```

```python
# OO method from scratch 
fig, ax = plt.subplots(figsize = (10, 6))
#plot the data 
scatter = ax.scatter(x= over_50['age'], y= over_50['chol'], c= over_50['target']);

#customzie the plot
ax.set(title= "Heart disease vs Cholesterol Level",xlabel="age",ylabel= 'Cholesterol');

#add a legend 
ax.legend(*scatter.legend_elements(),title="Target" );
#add horizonal line of plot 
ax.axhline(over_50['chol'].mean(),linestyle ='--');
```

```python
over_50.head()
```

```python
#Subplot of col,age and thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,figsize =(10,10))

#on ax0

#Add data to ax0
scatter  = ax0.scatter(x= over_50['age'],y=over_50['chol'],c=over_50['target'])
#customize ax0
ax0.set(title ='Heart Disease vs Cholesterol Level',xlabel='Age',ylabel= 'Cholesterol')
#add a legend 
ax0.legend(*scatter.legend_elements(),title="Target" )
#add horizonal line of plot 
ax0.axhline(over_50['chol'].mean(),linestyle ='--');

#on ax1


#Add data to ax0
scatter  = ax1.scatter(x= over_50['age'],y=over_50['thalach'],c=over_50['target'])
#customize ax0
ax1.set(title ='Heart Disease vs Max Heart Rate',xlabel='Age',ylabel= 'Max Heart Rate')
#add a legend 
ax1.legend(*scatter.legend_elements(),title="Target" )
#add horizonal line of plot 
ax1.axhline(over_50['thalach'].mean(),linestyle ='--');

```

```python
#We can reproduce above plotts by keeping X axis shared/ common 

#Subplot of col,age and thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,figsize =(10,10),sharex=True)

#on ax0

#Add data to ax0
scatter  = ax0.scatter(x= over_50['age'],y=over_50['chol'],c=over_50['target'])
#customize ax0
ax0.set(title ='Heart Disease vs Cholesterol Level',ylabel= 'Cholesterol')
#add a legend 
ax0.legend(*scatter.legend_elements(),title="Target" )
#add horizonal line of plot 
ax0.axhline(over_50['chol'].mean(),linestyle ='--');

#on ax1


#Add data to ax0
scatter  = ax1.scatter(x= over_50['age'],y=over_50['thalach'],c=over_50['target'])
#customize ax0
ax1.set(title ='Heart Disease vs Max Heart Rate',xlabel='Age',ylabel= 'Max Heart Rate')
#add a legend 
ax1.legend(*scatter.legend_elements(),title="Target" )
#add horizonal line of plot 
ax1.axhline(over_50['thalach'].mean(),linestyle ='--');

#add title to the figure 
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```

[matplotlib color scheme weblink](https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html)

```python
#Customize x and y axis limitation 




#Subplot of col,age and thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,figsize =(10,10),sharex=True)

#on ax0

#Add data to ax0
scatter  = ax0.scatter(x= over_50['age'],y=over_50['chol'],c=over_50['target'],cmap= 'winter')
#customize ax0
ax0.set(title ='Heart Disease vs Cholesterol Level',ylabel= 'Cholesterol')
#set x limit
ax0.set_xlim([50,80])
#add a legend 
ax0.legend(*scatter.legend_elements(),title="Target" )
#add horizonal line of plot 
ax0.axhline(over_50['chol'].mean(),linestyle ='--');

#on ax1


#Add data to ax0
scatter  = ax1.scatter(x= over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap= 'winter')
#customize ax0
ax1.set(title ='Heart Disease vs Max Heart Rate',xlabel='Age',ylabel= 'Max Heart Rate')
#set x limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#add a legend 
ax1.legend(*scatter.legend_elements(),title="Target" )
#add horizonal line of plot 
ax1.axhline(over_50['thalach'].mean(),linestyle ='--');

#add title to the figure 
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```
